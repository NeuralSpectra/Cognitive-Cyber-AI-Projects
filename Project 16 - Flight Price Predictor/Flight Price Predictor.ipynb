# **Model Building**

**X contains the selected features from combine_df for modeling.**

**y contains the target variable "Price" from combine_df.**

```python
X = combine_df[['Route', 'Date', 'Month', 'stop', 'Arrival_Hour',
       'Arrival_Minute', 'Departure_Hour', 'Departure_Minute', 'Air Asia',
       'Air India', 'GoAir', 'IndiGo', 'Jet Airways', 'Jet Airways Business',
       'Multiple carriers', 'Multiple carriers Premium economy', 'SpiceJet',
       'Trujet', 'Vistara', 'Vistara Premium economy', '1 Long layover',
       '1 Short layover', '2 Long layover', 'Business class',
       'Change airports', 'In-flight meal not included',
       'No check-in baggage included', 'No info', 'Red-eye flight', 'Banglore',
       'Chennai', 'Delhi', 'Kolkata', 'Mumbai', 'Destination_Banglore',
       'Destination_Cochin', 'Destination_Delhi', 'Destination_Hyderabad',
       'Destination_Kolkata', 'Destination_New Delhi']]

y = combine_df[["Price"]]
```

**This code snippet correctly scales the features (X) and the target (y) using MinMaxScaler from scikit-learn, ensuring both are scaled between 0 and 1.**

```python
from sklearn.preprocessing import MinMaxScaler

Scaler_X = MinMaxScaler().set_output(transform = "pandas")

X = Scaler_X.fit_transform(X)

Scaler_y = MinMaxScaler().set_output(transform = "pandas")

y = Scaler_y.fit_transform(y)
```

**These imports allow for splitting data into training and testing sets (train_test_split), and evaluating models using metrics like mean squared error (mean_squared_error) in scikit-learn.**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

**This command splits the dataset into training and testing sets (X_train, X_test for features and y_train, y_test for target) using 80% of the data for training and 20% for testing.**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.80)
```

**1: Import Model: Imports RandomForestRegressor for building a random forest ensemble.**

**2: Create Model (Depth 10): Creates a random forest regressor (RFR) with a maximum tree depth of 10 (controls model complexity).**

**3: Train Model: Trains the RFR model on X_train features and y_train targets.**

**4: Predict & Evaluate: Makes predictions on X_test, calculates MSE vs y_test, prints the error.**

```python
from sklearn.ensemble import RandomForestRegressor

RFR = RandomForestRegressor(max_depth=10)

RFR.fit(X_train, y_train)

y_pred = RFR.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)

print("Random Forest Mean Squared Error: ", MSE)
```

**1: Import Model: Imports DecisionTreeRegressor for creating a decision tree regression model.**

**2: Create Model (Split & Depth Control): Sets up DTR model with minimum samples required for a split set to 5 and maximum tree depth set to 20 (controls model complexity).**

**3: Train Model: Trains the DTR model on X_train features and y_train targets.**

**4: Predict & Evaluate: Makes predictions on X_test, calculates MSE vs y_test, prints the error.**

```python
from sklearn.tree import DecisionTreeRegressor

DTR = DecisionTreeRegressor(min_samples_split=5, max_depth=20)

DTR.fit(X_train, y_train)

y_pred = DTR.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)

print("Decision Tree Mean Squared Error: ", MSE)
```

**1: Import Model: Imports SVR from sklearn.svm for Support Vector Regression.**

**2: Create Model (Regularization Param): Creates an SVR model (svr) with a regularization parameter C set to 1.0 (controls model complexity).**

**3: Train Model: Trains the svr model on X_train features and y_train targets.**

**4: Predict & Evaluate: Makes predictions on X_test, calculates MSE vs y_test, prints the error.**

```python
from sklearn.svm import SVR

svr = SVR(C=1.0)

svr.fit(X_train, y_train)

y_pred = svr.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)

print("SVR Mean Squared Error: ", MSE)
```

**1: Import Model: Imports Ridge from sklearn.linear_model for ridge regression.**

**2: Create Model (Regularization Strength): Creates a ridge regression model (Rg) with an alpha value of 1.0 (controls the amount of L2 regularization, higher alpha means stronger regularization).**

**3: Train Model: Trains the Rg model on X_train features and y_train targets.**

**4: Predict & Evaluate: Makes predictions on X_test, calculates MSE vs y_test, prints the error.**

```python
from sklearn.linear_model import Ridge

Rg = Ridge(alpha=1.0)

Rg.fit(X_train, y_train)

y_pred = Rg.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)

print("Ridge Mean Squared Error: ", MSE)
```

**1: Import Model: Imports Lasso from sklearn.linear_model for L1-regularized linear regression.**

**2: Create Model (Regularization Strength): Creates a Lasso model (Ls) with an alpha value of 1.0 (controls the amount of L1 regularization, higher alpha leads to more feature coefficient shrinkage to zero).**

**3: Train Model: Trains the Ls model on X_train features and y_train targets.**

**4: Predict & Evaluate: Makes predictions on X_test, calculates MSE vs y_test, prints the error.**

```python
from sklearn.linear_model import Lasso

Ls = Lasso(alpha=1.0)

Ls.fit(X_train, y_train)

y_pred = Ls.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)

print("Lasso Mean Squared Error: ", MSE)
```
