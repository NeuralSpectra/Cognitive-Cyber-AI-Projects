# **Model Building**

**This code defines the feature matrix "X" containing the selected attributes for modeling, including player attributes, club information, nationality, and position. It also defines the target variable "y" containing the release clauses of players.**

```python
X = df[['Age', 'Overall', 'Potential', 'International Reputation', 'Weak Foot',
       'Skill Moves', 'Jersey Number', 'Contract Valid Until', 'Weight',
       'Crossing', 'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Dribbling',
       'LongPassing', 'BallControl', 'Acceleration', 'SprintSpeed', 'Agility',
       'Reactions', 'Balance', 'ShotPower', 'Stamina', 'Strength',
       'Interceptions', 'Positioning', 'Vision', 'Composure', 'Marking',
       'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes',
       'Club_Atlético Madrid', 'Club_Chelsea', 'Club_FC Barcelona',
       'Club_FC Bayern München', 'Club_Juventus', 'Club_Manchester City',
       'Club_Manchester United', 'Club_Paris Saint-Germain',
       'Club_Real Madrid', 'Club_Tottenham Hotspur', 'Nationality_Argentina',
       'Nationality_Belgium', 'Nationality_Brazil', 'Nationality_Croatia',
       'Nationality_England', 'Nationality_France', 'Nationality_Germany',
       'Nationality_Italy', 'Nationality_Poland', 'Nationality_Portugal',
       'Nationality_Slovenia', 'Nationality_Spain', 'Nationality_Uruguay',
       'Preferred Foot_Left', 'Preferred Foot_Right']]

y = df[['Release Clause']]
```

**The code you provided uses the MinMaxScaler from scikit-learn to scale features to a given range, typically between 0 and 1. The set_output(transform="pandas") method specifies that the output should be a pandas DataFrame.**

```python
from sklearn.preprocessing import MinMaxScaler

Scaler = MinMaxScaler().set_output(transform="pandas")

X = Scaler.fit_transform(X)

Scaler_y = MinMaxScaler().set_output(transform = "pandas")

y = Scaler_y.fit_transform(y)
```

**train_test_split: Used for splitting datasets into training and testing sets for model validation.**

**mean_squared_error: Computes the mean squared error between predicted and true values, commonly used to evaluate regression models.**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

**This code splits the feature matrix "X" and the target variable "y" into training and testing subsets using the "train_test_split()" function. It assigns 80% of the data to the training set and 20% to the testing set. The resulting subsets are named "X_train", "X_test", "y_train", and "y_test".**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

**1: Import the LinearRegression class from the sklearn.linear_model module.**

**2: Initialize a linear regression model instance named LR.**

**3: Train the linear regression model using the training data (X_train and y_train) with the fit() method.**

**4: Make predictions on the testing data using the trained model with the predict() method, assigning the predictions to y_pred.**

**5: mean_squared_error(y_test, y_pred) computes the squared average difference between predicted (y_pred) and actual (y_test) values, gauging model accuracy.**

**6: print("Linear Regression MSE:", MSE) outputs the MSE, indicating how well the linear regression model fits the test data, with smaller values signifying better performance.**







```python
from sklearn.linear_model import LinearRegression

LR  = LinearRegression()

LR.fit(X_train, y_train)

y_pred = LR.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)

print("Linear Regression Mean Squared Error:", MSE)
```

**1: Import the "DecisionTreeRegressor" class from the "sklearn.tree "module.**

**2: Instantiate a decision tree regressor named "DTR".**

**3: Train the decision tree regressor model using the training data ("X_train" and "y_train") with the "fit()" method.**

**4: Predict the release clauses for the testing data using the trained model with the "predict()" method, storing the predictions in "y_pred".**

**5: mean_squared_error(y_test, y_pred) computes the squared average difference between predicted (y_pred) and actual (y_test) values, gauging model accuracy.**

**6: print("Decision Tree MSE:", MSE) outputs the MSE, indicating how well the Decision Tree Regressor model fits the test data, with smaller values signifying better performance.**


```python
from sklearn.tree import DecisionTreeRegressor

DTR = DecisionTreeRegressor()

DTR.fit(X_train, y_train)

y_pred = DTR.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)

print("Decision Tree Mean Squared Error:", MSE)
```

**1: Import the RandomForestRegressor class from the sklearn.ensemble module.**

**2: Instantiate a random forest regressor named RFR with 1000 estimators and utilizing all available CPU cores for parallel processing.**

**3: Train the random forest regressor model using the training data (X_train and y_train) with the fit() method.**

**4: Predict the release clauses for the testing data using the trained model with the predict() method, storing the predictions in y_pred.**

**5: mean_squared_error(y_test, y_pred) computes the squared average difference between predicted (y_pred) and actual (y_test) values, gauging model accuracy.**

**6: print("Random Forest MSE:", MSE) outputs the MSE, indicating how well the Random Forest Regressor model fits the test data, with smaller values signifying better performance.**

```python
from sklearn.ensemble import RandomForestRegressor

RFR  = RandomForestRegressor(n_estimators=1000, n_jobs = -1)

RFR.fit(X_train, y_train)

y_pred = RFR.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)

print("Random Forest Mean Squared Error:", MSE)
```

**1: Import the Ridge class from the sklearn.linear_model module.**

**2: Instantiate a ridge regression model named Rg with a regularization strength (alpha) set to 0.1.**

**3: Train the ridge regression model using the training data (X_train and y_train) with the fit() method.**

**4: Predict the release clauses for the testing data using the trained model with the predict() method, storing the predictions in y_pred.**

**5: mean_squared_error(y_test, y_pred) computes the squared average difference between predicted (y_pred) and actual (y_test) values, gauging model accuracy.**

**6: print("Ridge MSE:", MSE) outputs the MSE, indicating how well the  Ridge model fits the test data, with smaller values signifying better performance.**

```python
from sklearn.linear_model import Ridge

Rg = Ridge(alpha=0.1)

Rg.fit(X_train, y_train)

y_pred = Rg.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)

print("Ridge Mean Squared Error:", MSE)
```

**1: Import the Lasso class from the sklearn.linear_model module.**

**2: Instantiate a Lasso regression model named Ls with a regularization strength (alpha) set to 0.1.**

**3: Train the Lasso regression model using the training data (X_train and y_train) with the fit() method.**

**4: Predict the release clauses for the testing data using the trained model with the predict() method, storing the predictions in y_pred.**

**5: mean_squared_error(y_test, y_pred) computes the squared average difference between predicted (y_pred) and actual (y_test) values, gauging model accuracy.**

**6: print("Lasso MSE:", MSE) outputs the MSE, indicating how well the Lasso model fits the test data, with smaller values signifying better performance.**

```python
from sklearn.linear_model import Lasso

Ls = Lasso(alpha=0.1)

Ls.fit(X_train, y_train)

y_pred = Ls.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)

print("Lasso Mean Squared Error:", MSE)
```
