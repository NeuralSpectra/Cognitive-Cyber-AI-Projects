# **Model Building**

**This code defines the features (X) and the target variable (y) for the dataset. The features include various attributes of the sculptures and transaction details, while the target variable is the "Cost" of the sculptures.**

```python
X = Artist[['Artist Reputation', 'Height', 'Width', 'Weight', 'Price Of Sculpture',
       'Base Shipping Price', 'Material_Aluminium', 'Material_Brass',
       'Material_Bronze', 'Material_Clay', 'Material_Marble', 'Material_Stone',
       'Material_Wood', 'International_No', 'International_Yes',
       'Express Shipment_No', 'Express Shipment_Yes',
       'Installation Included_No', 'Installation Included_Yes',
       'Transport_Airways', 'Transport_Roadways', 'Transport_Waterways',
       'Fragile_No', 'Fragile_Yes', 'Remote Location_No',
       'Remote Location_Yes', 'Customer Information_Wealthy',
       'Customer Information_Working Class']]

y = Artist["Cost"]
```

**This code generates a heatmap using Seaborn and Matplotlib libraries to visualize the correlation matrix of the DataFrame "X", with annotations, a Viridis colormap, and specified figure size.**


```python
plt.figure(figsize=(13,13))
sns.heatmap(X.corr(),annot=True,cmap='viridis',linewidths=.5)
```

**These imports from scikit-learn allow for splitting the dataset into training and testing sets and evaluating the accuracy of machine learning models.**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

**This code splits the features (X) and target variable (y) into training and testing sets using a train size of 80% for training and 20% for testing.**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

**1: Importing the RandomForestRegressor class from scikit-learn's ensemble module.**

**2: Initializing a Random Forest Regressor with 200 decision trees and parallelizing the process across all available CPU cores, then fitting the model with training data.**

**3: Using the trained model to predict target values for the test dataset.**

**4: Determining the accuracy of the model on the training data by computing the coefficient of determination (R^2) and converting it to a percentage.**

**5: Evaluating the model's performance on the test data using the same metric and printing both training and testing accuracies for comparison.**

```python
from sklearn.ensemble import RandomForestRegressor

RFR = RandomForestRegressor(n_estimators=200, n_jobs = -1)

RFR.fit(X_train, y_train)

y_pred = RFR.predict(X_test)

RFC_Training_Accuracy = round(RFR.score(X_train, y_train) * 100, 2)
print("Random Forest Training Accuracy: ", round(RFC_Training_Accuracy, 2), '%')

RFC_Testing_Accuracy = round(RFR.score(X_test, y_test) * 100, 2)
print("Random Forest Testing Accuracy: ", round(RFC_Testing_Accuracy, 2), '%')
```

**1: Importing the LinearRegression class from scikit-learn's linear_model module.**

**2: Creating an instance of the Linear Regression model and fitting it with training data (X_train, y_train).**

**3: Using the trained model to predict target values for the test dataset (X_test).**

**4: Determining the accuracy of the model on the training data by computing the coefficient of determination (R^2) and converting it to a percentage.**

**5: Evaluating the model's performance on the test data using the same metric and printing both training and testing accuracies for comparison.**

```python
from sklearn.linear_model import LinearRegression

LR = LinearRegression()

LR.fit(X_train, y_train)

y_pred = LR.predict(X_test)

LR_Training_Accuracy = round(LR.score(X_train, y_train) * 100, 2)
print("Linear Regression Training Accuracy: ", round(LR_Training_Accuracy, 2), '%')

LR_Testing_Accuracy = round(LR.score(X_test, y_test) * 100, 2)
print("Linear Regression Testing Accuracy: ", round(LR_Testing_Accuracy, 2), '%')
```

**1: Importing the DecisionTreeRegressor class from scikit-learn's tree module.**

**2: Model Initialization and Training: Creating an instance of the Decision Tree Regressor model and fitting it with training data (X_train, y_train).**

**3: Making Predictions: Using the trained model to predict target values for the test dataset (X_test).**

**4: Calculating Training Accuracy: Determining the accuracy of the model on the training data by computing the coefficient of determination (R^2) and converting it to a percentage.**

**5: Calculating Testing Accuracy: Evaluating the model's performance on the test data using the same metric and printing both training and testing accuracies for comparison.**

```python
from sklearn.tree import DecisionTreeRegressor

DTR = DecisionTreeRegressor()

DTR.fit(X_train, y_train)

y_pred = DTR.predict(X_test)

DTR_Training_Accuracy = round(DTR.score(X_train, y_train) * 100, 2)
print("Decision Tree Training Accuracy: ", round(DTR_Training_Accuracy, 2), "%")

DTR_Testing_Accuracy = round(DTR.score(X_test, y_test) * 100, 2)
print("Decision Tree Testing Accuracy: ", round(DTR_Testing_Accuracy, 2), "%")
```
