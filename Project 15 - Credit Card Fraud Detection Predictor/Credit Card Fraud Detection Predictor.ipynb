# **Model Building**

**X includes columns 'Time', 'V1' through 'V28', and 'Amount' from DataFrame df, serving as independent variables.**

**y consists of the 'Class' column from DataFrame df, representing the dependent variable indicating transaction fraudulence (1) or normalcy (0).**

```python
X = df[['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']]

y = df[["Class"]]
```

**MinMaxScaler Usage: MinMaxScaler from sklearn.preprocessing is employed to scale features (X) to a specified range, often [0, 1], ensuring numerical consistency across variables.**

**Scaling Features: scaler_X = MinMaxScaler().fit(X) initializes and fits the scaler to features (X), then transforms them (X_scaled = scaler_X.transform(X)) to a scaled format (X_scaled).**

**Scaling Target Variable: Similarly, scaler_y = MinMaxScaler().fit(y) initializes and fits another scaler for the target variable (y), transforming it (y_scaled = scaler_y.transform(y)) to maintain proportionality with the scaled features.**

```python
from sklearn.preprocessing import MinMaxScaler

scaler_X = MinMaxScaler().set_output(transform="pandas")

X = scaler_X.fit_transform(X)

scaler_y = MinMaxScaler().set_output(transform="pandas")

y = scaler_y.fit_transform(y)
```

**df["Class"].value_counts() provides a count of unique values in the 'Class' column of DataFrame df, indicating the number of transactions classified as normal (0) and fraudulent (1).**

```python
df["Class"].value_counts()
```

**andom Oversampling: RandomOverSampler from imblearn.over_sampling is used to balance class distribution by increasing instances of the minority class ("not majority").**

**Oversampling Process: ROS = RandomOverSampler(sampling_strategy="not majority") initializes oversampling strategy.**

**Visualization: y_ros.value_counts().plot.pie(autopct="%.2f") shows class distribution post-oversampling with a pie chart, displaying percentages for each class.**

**After performing Random Oversampling, the previously imbalanced data is now balanced, making it suitable for all types of classification models.**

```python
from imblearn.over_sampling import RandomOverSampler

ROS = RandomOverSampler(sampling_strategy="not majority")

X_ros, y_ros = ROS.fit_resample(X, y)

ax = y_ros.value_counts().plot.pie(autopct = "%.2f")
_ = ax.set_title("Class Distribution After Random Oversampling")
```

**y_ros.value_counts() computes and returns the count of each unique value in the array y_ros, which represents the target variable (y) after random oversampling. It shows the distribution of classes (fraudulent and non-fraudulent transactions) post-oversampling.**

```python
y_ros.value_counts()
```

**train_test_split: Function from sklearn.model_selection for splitting datasets into training and testing sets.**

**mean_squared_error: Metric from sklearn.metrics to calculate the mean squared error between predicted and actual values in regression tasks.**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

**train_test_split(X_ros, y_ros, train_size=0.80, random_state=42) splits the oversampled data (X_ros and y_ros) into training (X_train and y_train) and testing (X_test and y_test) sets, with 80% of the data allocated for training and 20% for testing, ensuring reproducibility with random_state=42.**

```python
X_train, X_test, y_train, y_test = train_test_split(X_ros, y_ros, train_size=0.80, random_state=42)
```

**1: Import Logistic Regression: Import the LogisticRegression class from the sklearn.linear_model module.**

**2: LogReg is an instance of LogisticRegression with max iterations set to 100.**

**3: Fit the Model: Train the model using the training data (X_train and y_train) with LogReg.fit(X_train, y_train). This step finds the optimal parameters of the logistic regression model.**

**4: The code generates a classification report for Logistic Regression, comparing the true labels (y_test) with the predicted labels (y_pred).**

```python
from sklearn.linear_model import LogisticRegression

LogReg = LogisticRegression(max_iter = 100)

LogReg.fit(X_train, y_train)

y_pred = LogReg.predict(X_test)

CR = classification_report(y_test, y_pred)
print("Logistic Regression Cassification Report: ")
print("")
print(CR)
```

**1: Import Random Forest Classifier: Import the RandomForestClassifier class from the sklearn.ensemble module.**

**2: RFC is an instance of RandomForestClassifier with a minimum samples split set to 10.**

**3: Fit the Model: Train the model using the training data (X_train and y_train) with RFR.fit(X_train, y_train). This step builds multiple decision trees based on different subsets of the data.**

**4: The code generates a classification report for the Random Forest Classifier, comparing the true labels (y_test) with the predicted labels (y_pred).**

```python
from sklearn.ensemble import RandomForestClassifier

RFR = RandomForestClassifier(min_samples_split = 10)

RFR.fit(X_train, y_train)

y_pred = RFR.predict(X_test)

CR = classification_report(y_test, y_pred)
print("Random Forest Classifier Cassification Report: ")
print("")
print(CR)
```

**1: Import Decision Tree Classifier: Import the DecisionTreeClassifier class from the sklearn.tree module.**

**2: DTC is an instance of DecisionTreeClassifier with a maximum depth set to 20.**

**3: Fit the Model: Train the model using the training data (X_train and y_train) with DTC.fit(X_train, y_train). This step builds a decision tree based on the training data.**

**4: The code generates a classification report for the Decision Tree Classifier, comparing the true labels (y_test) with the predicted labels (y_pred).**

```python
from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier(max_depth=20)

DTC.fit(X_train, y_train)

y_pred = DTC.predict(X_test)

CR = classification_report(y_test, y_pred)
print("Decision Tree Classifer Cassification Report: ")
print("")
print(CR)
```

Therefore, we choose LogReg as our best model due to its balance between high precision and lower risk of overfitting.**
