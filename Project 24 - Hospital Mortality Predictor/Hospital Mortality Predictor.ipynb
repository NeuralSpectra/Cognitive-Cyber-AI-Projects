# **Data Cleaning**

**Counts and displays the number of missing (null) values in each column of the DataFrame.**

```python
df.isnull().sum()
```

**Drops rows where the "outcome" column has missing values and then checks for any remaining missing values in the DataFrame.**

```python
df = df.dropna(subset=["outcome"])
df.isnull().sum()
```

**Imputes missing values in specified columns of the DataFrame with the mean of each column, and then checks for any remaining missing values.**

```python
from sklearn.impute import SimpleImputer

Impute = SimpleImputer(strategy = "mean")

df["BMI"] = Impute.fit_transform(df[["BMI"]])

df["heart rate"] = Impute.fit_transform(df[["heart rate"]])

df["Systolic blood pressure"] = Impute.fit_transform(df[["Systolic blood pressure"]])

df["Diastolic blood pressure"] = Impute.fit_transform(df[["Diastolic blood pressure"]])

df["Respiratory rate"] = Impute.fit_transform(df[["Respiratory rate"]])

df.isnull().sum()
```

**Imputes missing values in specified columns of the DataFrame using K-Nearest Neighbors (KNN) with 5 neighbors, and then checks for any remaining missing values.**

```python
from sklearn.impute import KNNImputer

KNN = KNNImputer(n_neighbors = 5)

df["temperature"] = KNN.fit_transform(df[["temperature"]])

df["SP O2"] = KNN.fit_transform(df[["SP O2"]])

df["Urine output"] = KNN.fit_transform(df[["Urine output"]])

df["Neutrophils"] = KNN.fit_transform(df[["Neutrophils"]])

df["Lymphocyte"] = KNN.fit_transform(df[["Lymphocyte"]])

df.isnull().sum()
```

**Fills missing values in specified columns of the DataFrame with the mean of each respective column and then checks for any remaining missing values.**

```python
df["Basophils"] = df["Basophils"].fillna(df["Basophils"].mean())

df["Lactic acid"] = df["Lactic acid"].fillna(df["Lactic acid"].mean())

df["PH"] = df["PH"].fillna(df["PH"].mean())

df["PCO2"] = df["PCO2"].fillna(df["PCO2"].mean())

df["Creatine kinase"] = df["Creatine kinase"].fillna(df["Creatine kinase"].mean())

df["PT"] = df["PT"].fillna(df["PT"].mean())

df["INR"] = df["INR"].fillna(df["INR"].mean())

df["glucose"] = df["glucose"].fillna(df["glucose"].mean())

df["Blood calcium"] = df["Blood calcium"].fillna(df["Blood calcium"].mean())

df.isnull().sum()
```

# **Model Building**

**Defines X as the feature set containing the specified columns and y as the target variable containing the "outcome" column.**

```python
X = df[['group', 'age', 'gendera', 'BMI', 'hypertensive',
       'atrialfibrillation', 'CHD with no MI', 'diabetes', 'deficiencyanemias',
       'depression', 'Hyperlipemia', 'Renal failure', 'COPD', 'heart rate',
       'Systolic blood pressure', 'Diastolic blood pressure',
       'Respiratory rate', 'temperature', 'SP O2', 'Urine output',
       'hematocrit', 'RBC', 'MCH', 'MCHC', 'MCV', 'RDW', 'Leucocyte',
       'Platelets', 'Neutrophils', 'Basophils', 'Lymphocyte', 'PT', 'INR',
       'NT-proBNP', 'Creatine kinase', 'Creatinine', 'Urea nitrogen',
       'glucose', 'Blood potassium', 'Blood sodium', 'Blood calcium',
       'Chloride', 'Anion gap', 'Magnesium ion', 'PH', 'Bicarbonate',
       'Lactic acid', 'PCO2', 'EF']]

y = df[["outcome"]]
```

**Scales the feature set X and target variable y to a range between 0 and 1 using MinMaxScaler, with the output format set to pandas DataFrames.**

```python
from sklearn.preprocessing import MinMaxScaler

Scaler_X = MinMaxScaler().set_output(transform = "pandas")

X = Scaler_X.fit_transform(X)

Scaler_y = MinMaxScaler().set_output(transform = "pandas")

y = Scaler_y.fit_transform(y)
```

**Imports train_test_split for splitting data into training and testing sets, and classification_report for evaluating the performance of classification models.**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

**Splits the scaled features and target data into training and testing sets, with 80% of the data used for training.**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.80)
```

**1: Initialize Model: LogisticRegression is set up with a regularization strength C=0.1 and a maximum of 1000 iterations.**

**2: Train Model: The model is trained on the training data (X_train, y_train).**

**3: Make Predictions: The trained model predicts the target variable on the test data (X_test).**

**4: Evaluate Performance: classification_report generates a detailed report on precision, recall, and F1-score for each class, providing insights into model performance.**

```python
from sklearn.linear_model import LogisticRegression

LogReg = LogisticRegression(C=0.1, max_iter=1000)

LogReg.fit(X_train, y_train)

y_pred = LogReg.predict(X_test)

CR = classification_report(y_test, y_pred)

print("Classification Report For Logistic Regression")
print("")
print(CR)
```

**1: Initialize Model: RandomForestClassifier is set up with 100 trees (n_estimators=100).**

**2: Train Model: The model is trained on the training data (X_train, y_train).**

**3: Make Predictions: The trained model predicts the target variable on the test data (X_test).**

**4: Evaluate Performance: classification_report provides a detailed summary of precision, recall, and F1-score for each class, showing the model's classification performance.**

```python
from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier(n_estimators=100)

RFC.fit(X_train, y_train)

y_pred = RFC.predict(X_test)

CR = classification_report(y_test, y_pred)

print("Classification Report For Random Forest Classifier")
print("")
print(CR)
```

**1: Initialize Model: DecisionTreeClassifier is configured with a maximum depth of 100 and a minimum of 10 samples per leaf.**

**2: Train Model: The model is trained using the training dataset (X_train, y_train).**

**3: Make Predictions: The trained model makes predictions on the test dataset (X_test).**

**4: Evaluate Performance: classification_report displays the precision, recall, and F1-score for each class, summarizing the classifier's performance.**

```python
from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier(max_depth=100, min_samples_leaf=10)

DTC.fit(X_train, y_train)

y_pred = DTC.predict(X_test)

CR = classification_report(y_test, y_pred)

print("Classification Report For Decision Tree Classifier")
print("")
print(CR)
```

**1: Initialize Model: SVC is set with C=100 for regularization strength and gamma=0.1 for the kernel function.**

**2: Train Model: The Support Vector Classifier is trained using the training dataset (X_train, y_train).**

**3: Make Predictions: Predictions are made on the test dataset (X_test).**

**4: Evaluate Performance: classification_report provides detailed metrics like precision, recall, and F1-score for evaluating the model's classification performance.**

```python
from sklearn.svm import SVC

svc = SVC(C=100, gamma=0.1)

svc.fit(X_train, y_train)

y_pred = svc.predict(X_test)

CR = classification_report(y_test, y_pred)

print("Classification Report For Support Vector Classifier")
print("")
print(CR)
```
