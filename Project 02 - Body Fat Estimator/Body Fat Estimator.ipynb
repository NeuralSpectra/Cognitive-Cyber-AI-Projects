# **Handling Outliers**

**This code calculates the upper and lower limits for outlier detection using the z-score method with a threshold of 4 standard deviations from the mean for each column in the DataFrame df, storing the upper limits in the list upper_limit and the lower limits in the list lower_limit.**

```python
upper_limit = []
lower_limit = []

for column in df.columns:
    mean = df[column].mean()
    std_dev = df[column].std()
    upper_limit.append(mean + 4 * std_dev)
    lower_limit.append(mean - 4 * std_dev)
```

**This code selects 14 body measurement features as input variables (X) and the BodyFat column as the target variable (y) for machine learning.**

```python
X = df[['Density', 'Age', 'Weight', 'Height', 'Neck', 'Chest',
       'Abdomen', 'Hip', 'Thigh', 'Knee', 'Ankle', 'Biceps', 'Forearm',
       'Wrist']]

y = df[['BodyFat']]
```

**This code imports the ExtraTreesRegressor from scikit-learn, initializes an instance named er, and fits it to the feature data X and target variable y, training the model to predict "BodyFat" using an ensemble of decision trees.**

**The code creates a pandas Series, series, representing feature importances computed by the trained ExtraTreesRegressor model er, with feature names as indices.**

**1: This code initializes a MinMaxScaler for features.**

**2: It scales the feature data to a 0-1 range.**

**3: A separate scaler is created for the target variable.**

**4: The target values are also normalized using min-max scaling.**

```python
from sklearn.preprocessing import MinMaxScaler

Scaler = MinMaxScaler().set_output(transform="pandas")

X_scaled = Scaler.fit_transform(X)

Scaler_y = MinMaxScaler().set_output(transform="pandas")

y_scaled = Scaler_y.fit_transform(y)
```

# **Model Building**

**This Python code imports modules for train-test splitting, linear regression, and mean squared error computation from the scikit-learn library.**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

**This line splits the scaled features and target into training and testing sets, allocating 80% of the data for training and the remaining 20% for testing.**

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, train_size=0.8)
```

**This code snippet performs linear regression modeling with the parameter "fit_intercept=True", fitting the model to the training data, making predictions on the test data, and then calculating the Mean Squared Error (MSE) between the predicted and actual values, printing the result.**

```python
LR = LinearRegression(fit_intercept=True)

LR.fit(X_train, y_train)

y_pred = LR.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)
print(f"Linear Regression Mean Squared Error: {MSE}")
```

**This code utilizes a Random Forest Regressor to model the data, adding parameters "n_estimators=500" and "max_depth=None", fitting it to the training set, predicting on the test set, and then calculating the Mean Squared Error (MSE) between the predicted and actual values, printing the result.**

```python
from sklearn.ensemble import RandomForestRegressor

RFR = RandomForestRegressor(n_estimators=500, max_depth=None)

RFR.fit(X_train, y_train)

y_pred = RFR.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)
print(f"Random Forest Mean Squared Error: {MSE}")
```

**This code employs a Decision Tree Regressor to model the data, using the parameters "max_depth=5" and "min_samples_split=10", fitting it to the training set, predicting on the test set, and then calculating the Mean Squared Error (MSE) between the predicted and actual values, printing the result.**

```python
from sklearn.tree import DecisionTreeRegressor

DTR = DecisionTreeRegressor(max_depth=5, min_samples_split=10)

DTR.fit(X_train, y_train)

y_pred = DTR.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)
print(f"Decision Tree Mean Squared Error: {MSE}")
```

**This code uses Ridge regression to model the data, using the parameter "alpha=1.0", fitting it to the training set, predicting on the test set, and then calculating the Mean Squared Error (MSE) between the predicted and actual values, printing the result.**

```python
from sklearn.linear_model import Ridge

Rg = Ridge(alpha = 1.0)

Rg.fit(X_train, y_train)

y_pred = Rg.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)
print(f"Ridge Mean Squared Error: {MSE}")
```

**This code employs Lasso regression to model the data, using the parameters "alpha=1.0" and "max_iter=1000", fitting it to the training set, predicting on the test set, and then calculating the Mean Squared Error (MSE) between the predicted and actual values, printing the result.**

```python
from sklearn.linear_model import Lasso

Ls = Lasso(alpha=1.0, max_iter=1000)

Ls.fit(X_train, y_train)

y_pred = Ls.predict(X_test)

MSE = mean_squared_error(y_test, y_pred)
print(f"Lasso Mean Squared Error: {MSE}")
```

# **Conclusion**


- **Linear Regression (LR) model performed the best with MSE: 0.0007.**

- **Random Forest Regressor (RFR) model had a slightly higher MSE: 0.0012.**

- **Decision Tree Regressor (DTR) model had a significantly higher MSE: 0.0019, indicating poorer performance compared to LR and RFR.**

- **Ridge Regression (Rg) and Lasso (Ls) models had even higher MSE values of 0.0016 and 0.0277 respectively, suggesting they didn't fit the data well.**

**The Linear Regression (LR) model outperformed the other models with the lowest MSE of 0.0007. Therefore, it appears that Linear Regression is the most suitable model for this dataset, offering the best predictive performance among the models tested.**
