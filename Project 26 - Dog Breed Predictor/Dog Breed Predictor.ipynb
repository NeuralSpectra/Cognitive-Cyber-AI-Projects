**Filters the DataFrame to include only rows where the "breed" column matches one of the values in CLASS_NAMES, then resets the index of the filtered DataFrame and displays the first few rows.**

```python
CLASS_NAMES = df["breed"].unique()
labels = df[df["breed"].isin(CLASS_NAMES)]
labels = labels.reset_index(drop = True)

# CLASS_NAMES = ['boston_bull', 'dingo', 'pekinese', 'bluetick', 'golden_retriever',
#        'bedlington_terrier', 'borzoi', 'basenji', 'scottish_deerhound',
#        'shetland_sheepdog', 'walker_hound', 'maltese_dog',
#        'norfolk_terrier', 'african_hunting_dog',
#        'wire-haired_fox_terrier', 'redbone', 'lakeland_terrier', 'boxer',
#        'doberman', 'otterhound', 'standard_schnauzer',
#        'irish_water_spaniel', 'black-and-tan_coonhound', 'cairn',
#        'affenpinscher', 'labrador_retriever', 'ibizan_hound',
#        'english_setter', 'weimaraner', 'giant_schnauzer', 'groenendael',
#        'dhole', 'toy_poodle', 'border_terrier', 'tibetan_terrier',
#        'norwegian_elkhound', 'shih-tzu', 'irish_terrier', 'kuvasz', 'bernese_mountain_dog']

# labels = df[(df['breed'].isin(CLASS_NAMES))]
# labels = labels.reset_index()
# labels.head()

CLASS_NAMES = ['bernese_mountain_dog', 'scottish_deerhound', 'maltese_dog']
labels = df[(df['breed'].isin(CLASS_NAMES))]
labels = labels.reset_index()
labels.head()
```

**1: Initialize X_data: Creates an empty NumPy array to store images, with dimensions (number of samples, 224, 224, 3).**

**2: One-Hot Encoding: Uses label_binarize to convert breed labels into one-hot encoded format.**

**3: Load and Preprocess Images: Iterates through each sample, loads the image, resizes it to 224x224, converts it to an array, normalizes pixel values, and stores it in X_data.**

**4: Print Shapes and Sizes: Displays the shapes and sizes of the X_data and Y_data arrays.**

```python
from sklearn.preprocessing import label_binarize
from tqdm import tqdm
from tensorflow.keras.preprocessing import image

X_data = np.zeros((len(labels), 224, 224, 3), dtype = "float32")

Y_data = label_binarize(labels["breed"], classes=CLASS_NAMES)

for i in tqdm(range(len(labels))):
    img_path = "/content/Dog Dataset/train/%s.jpg" % labels["id"][i]
    img = image.load_img(img_path, target_size=(224, 224))
    img = image.img_to_array(img)
    X_data[i] = img / 255.0


print("\n Train Image Shape: ", X_data.shape, "Size: {:,}".format(X_data.size))
print("One-Hot Encoded Output Shape: ", Y_data.shape, "Size: {:,}".format(Y_data.size))
```

**Defines and compiles a Convolutional Neural Network (CNN) model with the following architecture:**

**Convolutional Layers:**

- **64 filters with 5*5 kernel, ReLU activation.**

- **32 filters with 3*5 kernel, ReLU activation, L2 regularization.**

- **16 filters with 7*7 kernel, ReLU activation, L2 regularization.**

- **8 filters with 5*5* kernel, ReLU activation, L2 regularization.**

**Pooling Layers:**

- **Max pooling with 2*2 pool size after each convolutional layer.**

**Fully Connected Layers:**

- **Dense layer with 128 units, ReLU activation, L2 regularization.**

- **Dense layer with 64 units, ReLU activation, L2 regularization.**

- **Output Dense layer with units equal to the number of classes (len(CLASS_NAMES)), softmax activation for classification.**

**Compilation:**

- **Loss function: Categorical Crossentropy.**

- **Optimizer: Adam with learning rate of 0.0001.**

- **Metric: Accuracy.**

**The model summary displays the architecture, number of parameters, and layer details.**

```python
from keras.models import Sequential
from keras.layers import Conv2D, Flatten, Dense, MaxPool2D, Dropout
from keras import regularizers
from tensorflow.keras.optimizers import Adam


model = Sequential()

model.add(Conv2D(filters = 64, kernel_size = (5,5), activation ='relu', input_shape = (224,224,3)))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(filters = 32, kernel_size = (3,3), activation ='relu', kernel_regularizer = 'l2'))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(filters = 16, kernel_size = (7,7), activation ='relu', kernel_regularizer = 'l2'))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(filters = 8, kernel_size = (5,5), activation ='relu', kernel_regularizer = 'l2'))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(128, activation = "relu", kernel_regularizer = 'l2'))
model.add(Dense(64, activation = "relu", kernel_regularizer = 'l2'))
model.add(Dense(len(CLASS_NAMES), activation = "softmax"))

model.compile(loss = 'categorical_crossentropy', optimizer = Adam(0.0001),metrics=['accuracy'])

model.summary()
```

**Split into Training/Validation and Test Sets:**

- **X_train_and_val, Y_train_and_val: Combined training and validation data.**
- **X_test, Y_test: Test data.**
- **Test set size is 10% of the total data.**

**Further Split Training/Validation Data:**

- **X_train, Y_train: Training data.**
- **X_val, Y_val: Validation data.**
- **Validation set size is 20% of the remaining data after the initial split.**

```python
from sklearn.model_selection import train_test_split

X_train_and_val, X_test, Y_train_and_val, Y_test = train_test_split(X_data, Y_data, test_size=0.1)

X_train, X_val, Y_train, Y_val = train_test_split(X_train_and_val, Y_train_and_val, test_size=0.2)
```

**It will output the dimensions of the arrays used for training, validation, and testing, which helps ensure the data has been split correctly.**

```python
print(f"X_train shape: {X_train.shape}")
print(f"Y_train shape: {Y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"Y_test shape: {Y_test.shape}")
print(f"X_val shape: {X_val.shape}")
print(f"Y_val shape: {Y_val.shape}")
```

**Trains the model with the following parameters:**

- **Training Data: X_train, Y_train**
- **Batch Size: 128**
- **Epochs: 100**
- **Validation Data: X_val, Y_val**

**The models performance on the validation set is monitored during training.**

```python
history = model.fit(X_train, Y_train, batch_size=128, epochs=100, validation_data=(X_val, Y_val))
```

**This code plots the training and validation accuracy over epochs:**

**1. Figure Size : Sets the size of the plot to 12x5 inches.**

**2. Plot Training Accuracy: Plots "istory.history["accuracy"]" in red.**

**3. Plot Validation Accuracy: Plots "history.history["val_accuracy"]' in blue.**

**4. Title and Labels : Adds a title, and labels for the x-axis and y-axis.**

**5. Legend : Differentiates between training and validation accuracies.**

**6. Show Plot : Displays the plot.**

**This visualization helps in understanding how well the model is performing over the training epochs.**

```python
plt.figure(figsize=(12, 5))
plt.plot(history.history["accuracy"], color="r")
plt.plot(history.history["val_accuracy"], color="b")
plt.title("Model Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend(["Train", "Validation"])
plt.show()
```

**1: Predict: Y_pred = model.predict(X_test) generates predictions for the test set.**

**2: Evaluate: score = model.evaluate(X_test, Y_test) computes the loss and accuracy on the test set.**

**3: Print Accuracy: Displays the test set accuracy as a percentage.**

**4: The score[1] represents the accuracy of the model on the test data.**

```python
Y_pred = model.predict(X_test)
score = model.evaluate(X_test, Y_test)
print("Accuracy Over The Test Set: \n", round((score[1] * 100), 2), "%")
```
