# **Data Cleaning**

* **Initializes separate "LabelEncoder" instances for "Ad Topic Line", "City", and "Country".**
* **Transforms the categorical columns in "df" into numerical labels using ".fit_transform()".**
* **Stores the transformed values back into the original DataFrame columns.**
* **Collects all encoders into a dictionary "label_encoders" for future use or inverse transformations.**

```python
from sklearn.preprocessing import LabelEncoder

le_ad_topic = LabelEncoder()
le_city = LabelEncoder()
le_country = LabelEncoder()

df["Ad Topic Line"] = le_ad_topic.fit_transform(df["Ad Topic Line"])
df["City"] = le_city.fit_transform(df["City"])
df["Country"] = le_country.fit_transform(df["Country"])

df.head(5)

label_encoders = {
    'Ad Topic Line': le_ad_topic,
    'City': le_city,
    'Country': le_country
}
```

**The code snippet processes the Timestamp column to extract individual time components. Hereâ€™s what each line does:**

**1: Convert to Datetime: df["Timestamp"] is converted to a datetime object**.

**2: Extract Year: Creates a new column Year with the year part of the timestamp.**

**3: Extract Month: Creates a new column Month with the month part of the timestamp.**

**4: Extract Day: Creates a new column Day with the day part of the timestamp.**

**5: Extract Hour: Creates a new column Hour with the hour part of the timestamp.**

**6: Extract Minute: Creates a new column Minute with the minute part of the
timestamp.**

**7: Extract Second: Creates a new column Second with the second part of the timestamp.**

```python
df["Timestamp"] = pd.to_datetime(df["Timestamp"])

df['Year'] = df['Timestamp'].dt.year
df['Month'] = df['Timestamp'].dt.month
df["Day"] = df['Timestamp'].dt.day
df["Hour"] = df['Timestamp'].dt.hour
df["Minute"] = df['Timestamp'].dt.minute
df["Second"] = df['Timestamp'].dt.second

df.head(5)
```

**This cleans up the DataFrame by removing the original timestamp data, leaving just the extracted time components and other relevant columns.**

```python
df.drop(columns=["Timestamp"], inplace=True)
df.head(5)
```

# **Model Building**

**X: Defines the feature set for the model, including numerical and encoded categorical data.**

**y: Defines the target variable, which is the binary outcome of whether the ad was clicked.**

```python
X = df[['Daily Time Spent on Site', 'Age', 'Area Income',
       'Daily Internet Usage', 'City', 'Male', 'Country',
        'Year', 'Month', 'Day', 'Hour', 'Minute', 'Second', 'Ad Topic Line']]

y = df[['Clicked on Ad']]
```

**train_test_split: Splits the data into training and testing sets to evaluate model performance.**

**classification_report: Generates a report with precision, recall, f1-score, and support for each class in classification problems.**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

**This function splits the dataset into training and testing sets with 80% of the data used for training and 20% for testing.**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.80)
```

**1: Scaler_X = MinMaxScaler().set_output(transform="pandas"): Initializes a Min-Max Scaler for features and sets the output format to a pandas DataFrame.**

**2: X = Scaler_X.fit_transform(X): Fits the scaler to the feature data (X) and transforms it to a scale between 0 and 1.**

**3: Scaler_y = MinMaxScaler().set_output(transform="pandas"): Initializes another Min-Max Scaler for the target variable and sets the output format to a pandas DataFrame.**

**4: y = Scaler_y.fit_transform(y): Fits the scaler to the target data (y) and transforms it to a scale between 0 and 1.**

```python
from sklearn.preprocessing import MinMaxScaler

Scaler_X = MinMaxScaler().set_output(transform = "pandas")

X = Scaler_X.fit_transform(X)

Scaler_y = MinMaxScaler().set_output(transform = "pandas")

y = Scaler_y.fit_transform(y)
```

**1: LogReg = LogisticRegression(max_iter=1000, fit_intercept=True): Initializes the Logistic Regression model with up to 1000 iterations and an intercept.**

**2: LogReg.fit(X_train, y_train): Trains the model on the training dataset.**

**3: y_pred = LogReg.predict(X_test): Predicts outcomes for the test dataset using the trained model.**

**4: CR = classification_report(y_test, y_pred): Produces a report showing the classification metrics for the model's predictions.**

```python
from sklearn.linear_model import LogisticRegression

LogReg = LogisticRegression(max_iter=1000, fit_intercept=True)

LogReg.fit(X_train, y_train)

y_pred = LogReg.predict(X_test)

CR = classification_report(y_test, y_pred)

print("Cassification Report For LogisticRegression:")
print("")
print(CR)
```

**1: RFC = RandomForestClassifier(min_samples_split=10, max_samples=100, n_estimators=100): Initializes the Random Forest Classifier with specific parameters for minimum samples per split, maximum samples per tree, and number of trees.**

**2: RFC.fit(X_train, y_train): Trains the Random Forest model using the provided training data.**

**3: y_pred = RFC.predict(X_test): Uses the trained model to make predictions on the test data.**

**4: CR = classification_report(y_test, y_pred): Generates a detailed classification report that includes precision, recall, and F1-score for the modelâ€™s predictions.**

```python
from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier(min_samples_split=10, max_samples=100, n_estimators=100)

RFC.fit(X_train, y_train)

y_pred = RFC.predict(X_test)

CR = classification_report(y_test, y_pred)

print("Classification Report For Random Forest Classifier:")
print("")
print(CR)
```

**1: DTC = DecisionTreeClassifier(min_samples_leaf=10, max_depth=500): Initializes the Decision Tree Classifier with a minimum number of samples per leaf and a maximum depth for the tree.**

**2: DTC.fit(X_train, y_train): Trains the Decision Tree model using the training data.**

**3: y_pred = DTC.predict(X_test): Predicts the class labels for the test data using the trained model.**

**4: CR = classification_report(y_test, y_pred): Produces a classification report with metrics like precision, recall, and F1-score for the Decision Tree modelâ€™s predictions.**

```python
from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier(min_samples_leaf=10, max_depth=500)

DTC.fit(X_train, y_train)

y_pred = DTC.predict(X_test)

CR = classification_report(y_test, y_pred)

print("Classification Report For Decision Tree Classifier:")
print("")
print(CR)
```

**1: svc = SVC(random_state=42, C=1.0): Initializes the Support Vector Classifier with a specified regularization parameter and random state for reproducibility.**

**2: svc.fit(X_train, y_train): Trains the Support Vector Machine model using the training data.**

**3: y_pred = svc.predict(X_test): Predicts the class labels for the test data using the trained SVM model.**

**4: CR = classification_report(y_test, y_pred): Generates a classification report with metrics like precision, recall, and F1-score for the SVM modelâ€™s predictions.**

```python
from sklearn.svm import SVC

svc = SVC(random_state=42, C=1.0)

svc.fit(X_train, y_train)

y_pred = svc.predict(X_test)

CR = classification_report(y_test, y_pred)

print("Classification Report For Support Vecto Machine:")
print("")
print(CR)
```
