**This function splitSequence takes a sequence seq and a number of time steps n_steps as input, then splits the sequence into input (X) and output (y) pairs for a supervised learning problem, where X contains sequences of length n_steps and y contains the corresponding next element in the sequence. It returns X and y as numpy arrays.**

```python
def splitSequence(seq, n_steps):

  X = []
  y = []

  for i in range(len(seq)):
    lastIndex = i + n_steps

    if lastIndex > len(seq) - 1:
      break

    seq_X, seq_y = seq[i:lastIndex], seq[lastIndex]

    X.append(seq_X)
    y.append(seq_y)
    pass

  X = np.array(X)
  y = np.array(y)

  return X, y
```

**This code snippet calls the splitSequence function with the data sequence and a value of n_steps equal to 7. It splits the data sequence into input-output pairs, where X contains sequences of 7 elements each, and y contains the next element in the sequence.**

```python
n_steps = 7
X, y = splitSequence(data, n_steps)
```

**Printing X would show the input sequences created by splitting the data sequence using a window size of 7. Each row represents a sequence of 7 consecutive elements from the original data sequence.**

```python
print(X)
```

**Printing y would display the corresponding output values, which are the next elements in the sequence following each sequence of 7 elements in X.**

```python
print(y)
```

**This code sets the number of features per time step to 1 (n_features = 1) and reshapes the input data X to have three dimensions: the number of samples, the number of time steps, and the number of features. It then prints the first two samples of the reshaped X.**

```python
n_features = 1
X = X.reshape((X.shape[0], X.shape[1], n_features))
print(X[:2])
```

**This code defines a Sequential model using Keras API from TensorFlow. It consists of an LSTM layer with 500 units and ReLU activation function, taking input of shape (n_steps, n_features), followed by a Dense layer with 1 unit, representing the output layer.**

```python
model = tf.keras.Sequential()
model.add(layers.LSTM(500, activation="relu", input_shape=(n_steps, n_features)))
model.add(layers.Dense(1))
```

**The model.summary() method provides a summary of the defined neural network model, including the type of each layer, its output shape, and the number of parameters.**

```python
model.summary()
```

# **Model Building**

**This code snippet utilizes scikit-learn's train_test_split function to split the input data (X) and target data (y) into training and testing sets. It allocates 20% of the data for testing (test_size=0.2) and assigns the remaining 80% to training.**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

**This code sets up the optimizer for the neural network model using the Adam optimizer with a learning rate of 0.001. It then compiles the model using the specified optimizer and loss function, which in this case is mean squared error (MSE), suitable for regression problems.**

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mean_squared_error')
```

**This code fits the compiled model to the training data. It specifies parameters such as batch size (64), number of epochs (100), and validation data (X_test and y_test). The training progress is displayed (verbose=1).**

```python
print("Fit Model On Training Data")
history = model.fit(
    X_train,
    y_train,
    batch_size = 64,
    epochs = 100,
    validation_data = (X_test, y_test), verbose=1
)
```

**This code prepares test data by creating a numpy array containing a sequence,
reshaping it to match the input shape expected by the model (1, n_steps, n_features).**

```python
test_data = np.array([5, 10, 15, 20, 25, 30, 35])
test_data = test_data.reshape((1, n_steps, n_features))
test_data
```

**This code uses the trained model to predict the next number in the sequence provided as test_data. The verbose=1 argument is used to display the progress of the prediction. The predicted next number is then printed.**

```python
predictNextNumber = model.predict(test_data, verbose=1)
print(predictNextNumber)
```

**This line saves a machine learning model trained on number sequences to a file named "Number-Sequence-Model.h5".**

**This code creates a figure to display the training and validation loss over epochs. It uses Matplotlib to plot the training and validation loss stored in the history object. The title and labels are set accordingly, and the plot is displayed.**

```python
fig = plt.figure()
fig.suptitle("Adam, lr=0.001, One Hidden Layer")
ax = fig.add_subplot(1, 1, 1)
ax.set_title("Cost")
ax.plot(history.history["loss"], label="Training")
ax.plot(history.history["val_loss"], label="Validation")
ax.legend()
plt.show()
```
