# **Model Building**

**This code separates the target variable "CLASS_LABEL" from the features in the DataFrame "Subset_Fishing", assigning the target variable to "y" and the remaining features to "X".**

```python
y = Subset_Fishing["CLASS_LABEL"]
X = Subset_Fishing.drop(["CLASS_LABEL"], axis = 1)
```

**This code imports necessary modules for machine learning tasks from scikit-learn, including "train_test_split" for splitting data, "accuracy_score" for evaluating model performance, and "RandomForestClassifier" for implementing a random forest classifier.**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

**This code splits the features (X) and target variable (y) into training and testing sets using the "train_test_split" function, with 80% of the data used for training ("train_size=0.8") and a specified random state for reproducibility ("random_state=42").**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

**1. Initialize Random Forest Classifier with 250 estimators and parallel processing enabled.**

**2. Fit the classifier to the training data and make predictions on the test data.**

**3. Calculate and print the rounded accuracy score on the training data.**

```python
RFC = RandomForestClassifier(n_estimators=250, n_jobs = -1)

RFC.fit(X_train, y_train)

y_pred = RFC.predict(X_test)

Random_Forest_Accuracy_Train = round(RFC.score(X_train, y_train) * 100, 2)

print(round(Random_Forest_Accuracy_Train, 2), "%")
```

**This code calculates the accuracy of the Random Forest Classifier on the test data ("X_test", "y_test") and prints the rounded accuracy score.**

```python
Random_Forest_Accuracy_Test = round(RFC.score(X_test, y_test) * 100, 2)
print(round(Random_Forest_Accuracy_Test, 2), "%")
```

**1. Import Logistic Regression classifier and initialize it with parameters: C=1.0, solver='liblinear'.**

**2. Fit the classifier to training data and make predictions on the test data.**

**3. Calculate and print the rounded accuracy score on the training data.**

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression(C=1.0, solver='liblinear')

LR.fit(X_train, y_train)

y_pred = LR.predict(X_test)

Logistic_Regression_Accuracy_Train = round(LR.score(X_train, y_train) * 100, 2)

print(round(Logistic_Regression_Accuracy_Train, 2), "%")
```

**This code calculates the accuracy of the Logistic Regression classifier on the test data ("X_test", "y_test") and prints the rounded accuracy score.**

```python
Logistic_Regression_Accuracy_Test = round(LR.score(X_test, y_test) * 100, 2)
print(round(Logistic_Regression_Accuracy_Test, 2), "%")
```

**1. Import Decision Tree Classifier and initialize it with parameters: max_depth=10, min_samples_split=2.**

**2. Fit the classifier to training data and make predictions on the test data.**

**3. Calculate and print the rounded accuracy score on the training data.**

```python
from sklearn.tree import DecisionTreeClassifier

DTC = DecisionTreeClassifier(max_depth=10, min_samples_split=2)

DTC.fit(X_train, y_train)

y_pred = DTC.predict(X_test)

Decision_Tree_Accuracy_Train = round(DTC.score(X_train, y_train) * 100, 2)

print(round(Decision_Tree_Accuracy_Train, 2), "%")
```

**This code calculates the accuracy of the Decision Tree classifier on the test data ("X_test", "y_test") and prints the rounded accuracy score.**

```python
Decision_Tree_Accuracy_Test = round(DTC.score(X_test, y_test) * 100, 2)
print(round(Decision_Tree_Accuracy_Test, 2), "%")
```
